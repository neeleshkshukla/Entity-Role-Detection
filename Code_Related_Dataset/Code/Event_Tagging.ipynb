```python
import os
import shutil

content_dir = '../../Data/input/content'
tag_dir = '../../Data/input/new_tags'
data_dir = '../../Data'

tag_file_content = ''

for f in os.listdir(os.path.join(content_dir)):
    tag_file_content = tag_file_content +f+','
    tag_file_content = tag_file_content +  f[:(f.index('_') + f[f.index('_')+1:].index('_'))+1] + '\n'
    
    
with open(os.path.join(data_dir, 'event_tag.csv'), 'w') as fw:
    fw.write(tag_file_content)
```

```python
with open(os.path.join(data_dir, 'event_tag.csv'), 'r') as f:
    for line in f:
        line = line.strip('\n')
        print(line.split(','))
```
