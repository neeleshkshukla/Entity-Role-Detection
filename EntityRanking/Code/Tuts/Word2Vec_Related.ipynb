#### Loading embeddings from txt file using genesim

```python
from gensim.models import KeyedVectors
```

```python
train_word_vectors = KeyedVectors.load_word2vec_format('../../../../trained_word_embeddings/word2vec/word_trained_from_corpus/w2v_corpus_trained_gensim_300.txt', binary=False)
print(train_word_vectors['was'])
```

```python
# No Need to run. Already converted
from gensim.scripts.glove2word2vec import glove2word2vec
glove_file = '../../../../pretrained_word_embeddings/Glove/glove.6B/glove.6B.300d.txt'
word2vec_file = '../../../../pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt'
glove2word2vec(glove_file, word2vec_file)
```

```python
pretrain_word_glove_vectors = KeyedVectors.load_word2vec_format('../../../../pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt', binary=False)
print(pretrain_word_glove_vectors['was'])
```

```python
train_pretrain_word_vectors = KeyedVectors.load_word2vec_format('../../../../trained_word_embeddings/word2vec/word_pretrain_trained_on_corpus/w2v_pretain_corpus_trained_gensim_300.txt', binary=False)
print(train_pretrain_word_vectors['was'])
```

```python
import scipy

train_word_vec_coprus_was = train_word_vectors['location']
train_word_vec_coprus_were = train_word_vectors['person']
pretrain_word_was = pretrain_word_glove_vectors['location']
pretrain_word_were = pretrain_word_glove_vectors['person']
train_pretrain_word_was = train_pretrain_word_vectors['location']
train_pretrain_word_were = train_pretrain_word_vectors['person']


dist_preTrain = scipy.spatial.distance.cosine(pretrain_word_was, pretrain_word_were)
print(dist_preTrain)
dist_corpTrain = scipy.spatial.distance.cosine(train_word_vec_coprus_was, train_word_vec_coprus_were)
print(dist_corpTrain)
dist_preTrainCorpTrain = scipy.spatial.distance.cosine(train_pretrain_word_was, train_pretrain_word_were)
print(dist_preTrainCorpTrain)
```
