```python
import os
import shutil
```

```python
content_dir = '../../../Data/input/content'
tag_dir = '../../../Data/input/new_tags'
target_dir = '../../../Data/output/Entity_Ranking/TypeRep/type_word/corpus_replaced_entity_with_tag'
_tag = 'PER_Accused'
```

```python
tag_word_content = ''

for f in os.listdir(content_dir):
    print(f)
    content_file = os.path.join(content_dir, f)
    tag_file = os.path.join(tag_dir, f)
    words = list()
    tags = list()
    with open(content_file, 'rt') as cf:
        for line in cf:
            for word in line.strip('\n').split():
                words.append(word)

    with open(tag_file, 'rt') as tf:
        for line in tf:
            for tag in line.strip('\n').split():
                tags.append(tag)
    word_len = len(words)
    tag_len = len(tags)
    if word_len == tag_len:
        i = 0
        while i < word_len:
            if tags[i] == _tag:
                entity_start_index = i
                entity_end_index = i 
                while entity_end_index< word_len -1 and tags[entity_end_index] == tags[entity_end_index+1]:
                    entity_end_index = entity_end_index + 1
                i = entity_end_index;
                tag_word_content = tag_word_content + _tag + ' '
            else:
                tag_word_content = tag_word_content + words[i] + ' '
            i = i + 1
            
with open(os.path.join(target_dir, _tag), "w") as text_file:
    text_file.write(tag_word_content)
```

```python
import gensim, logging
from gensim.models import Word2Vec, Phrases, phrases, KeyedVectors

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
class Corpus_Sentences(object):
    def __init__(self, filename):
        self.filename = filename;
        
    def __iter__(self):
        for line in open(self.filename):
            yield line.split()
                
sentences = Corpus_Sentences(os.path.join(target_dir, _tag)) # a memory-friendly iterator
```

```python
glove_wv = KeyedVectors.load_word2vec_format('../../../word_embeddings/pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt', binary=False)
```

```python
model = Word2Vec(size=300, min_count=1, iter=10)
model.build_vocab(sentences)
training_examples_count = model.corpus_count
# below line will make it 1, so saving it before
model.build_vocab([list(glove_wv.vocab.keys())], update=True)
```

```python
model.intersect_word2vec_format("../../../word_embeddings/pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt",binary=False, lockf=1.0)
```

```python
model.train(sentences, total_examples=training_examples_count, epochs=100)
print(model[_tag])
```

```python
model.similar_by_vector(model[_tag], topn=50)
```
