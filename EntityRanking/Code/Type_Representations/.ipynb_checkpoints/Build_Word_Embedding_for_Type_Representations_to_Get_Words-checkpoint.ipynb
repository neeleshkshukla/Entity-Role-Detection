```python
import gensim, logging, os
from gensim.models import Word2Vec, Phrases, phrases, KeyedVectors
import scipy, pickle


logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
target_dir = '../../../Data/output/Entity_Ranking/TypeRep/type_word/Type_Words'
source_dir = '../../../Data/output/Entity_Ranking/TypeRep/type_word/corpus_replaced_entity_with_tag'

_tag_word_dict = dict()
_tag_list = ['LOC_Event', 'LOC_Accused', 'LOC_Victim', 'LOC_Others', 'PER_Accused', 'PER_Victim', 'PER_Others', 'ORG_Accused', 'ORG_Victim', 'ORG_Others']
#_tag = 'LOC_Accused'
#_tag = 'LOC_Victim'
#_tag = 'LOC_Others'
#_tag = 'PER_Accused'
#_tag = 'PER_Victim'
#_tag = 'PER_Others'
#_tag = 'ORG_Accused'
#_tag = 'ORG_Victim'
#_tag = 'ORG_Others'
```

```python
class Corpus_Sentences(object):
    def __init__(self, filename):
        self.filename = filename;
        
    def __iter__(self):
        for line in open(self.filename):
            yield line.split()
```

```python
glove_wv = KeyedVectors.load_word2vec_format('../../../word_embeddings/pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt', binary=False)
```

```python
for _tag in _tag_list:
    sentences = Corpus_Sentences(os.path.join(source_dir, _tag)) # a memory-friendly iterator
    model = Word2Vec(size=300, min_count=1, iter=10)
    model.build_vocab(sentences)
    training_examples_count = model.corpus_count
    # below line will make it 1, so saving it before
    model.build_vocab([list(glove_wv.vocab.keys())], update=True)
    model.intersect_word2vec_format("../../../word_embeddings/pretrained_word_embeddings/Word2Vec/Glove_Converted/glove_2_word2vec.6B.300d.txt",binary=False, lockf=1.0)
    model.train(sentences, total_examples=training_examples_count, epochs=100)
    model.wv.save_word2vec_format(os.path.join('../../../word_embeddings/trained_word_embeddings/word2vec/type_rep_pretrain_trained_on_corpus', _tag), binary=False)
    similar_words = model.similar_by_word(_tag, topn=30)
    _tag_word_dict[_tag] = similar_words
    
pickle.dump(_tag_word_dict, open(os.path.join(target_dir, 'tag_word.p'), 'wb'))
```
