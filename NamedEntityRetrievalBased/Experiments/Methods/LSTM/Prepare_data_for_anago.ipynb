```python
import os
import numpy as np
```

```python
doc_dir = '../../../../Data/input/rectified/content'
anago = False # If anago create tsv file
guilla = True # if guillaumegenthial is used


train_file_list = []
valid_file_list = []
test_file_list = []

for f in os.listdir(doc_dir):
    #Random Sampling
    num = np.random.uniform(0,1)
    if num < 0.7:
        train_file_list.append(f)
    elif num < 0.9:
        valid_file_list.append(f)
    else:
        test_file_list.append(f)

print (len(train_file_list))
print (len(valid_file_list))
print (len(test_file_list))
```

```python
tag_dir = '../../../../Data/input/rectified/new_tags'
train_file_path = 'Data/train.txt'
valid_file_path = 'Data/valid.txt'
test_file_path = 'Data/test.txt'
max_sentence_length = 20

train_file = open(train_file_path, 'w')

for f in train_file_list:
    word_file_path = os.path.join(doc_dir, f)
    tag_file_path = os.path.join(tag_dir, f)
    lines_in_word_file = []
    lines_in_tag_file = []
    with open(word_file_path, "rt") as word_file:
        for line in word_file:
            lines_in_word_file.append(line)
    with open(tag_file_path, "rt") as tag_file:
        for line in tag_file:
            lines_in_tag_file.append(line)
    if (len(lines_in_word_file) == len(lines_in_tag_file)) and len(lines_in_word_file) > 0:
        for i in range(len(lines_in_word_file)):
            word_in_file = lines_in_word_file[i].split()
            tag_in_file = lines_in_tag_file[i].split()
            length = min(len(word_in_file), len(tag_in_file))
            #Create the word_tag pair
            sent_len = 0
            for j in range(length):
                if anago:
                    train_file.write(str(word_in_file[j])+'\t'+tag_in_file[j]+'\n')
                if guilla:
                    train_file.write(str(word_in_file[j])+' '+tag_in_file[j]+'\n')
                sent_len = sent_len + 1
                #if sent_len == max_sentence_length:
                    #sent_len = 0
                if word_in_file[j] == '.' and tag_in_file[j] == 'O':
                    train_file.write('\n')
train_file.close()

valid_file = open(valid_file_path, 'w')

for f in valid_file_list:
    word_file_path = os.path.join(doc_dir, f)
    tag_file_path = os.path.join(tag_dir, f)
    lines_in_word_file = []
    lines_in_tag_file = []
    with open(word_file_path, "rt") as word_file:
        for line in word_file:
            lines_in_word_file.append(line)
    with open(tag_file_path, "rt") as tag_file:
        for line in tag_file:
            lines_in_tag_file.append(line)
    if (len(lines_in_word_file) == len(lines_in_tag_file)) and len(lines_in_word_file) > 0:
        for i in range(len(lines_in_word_file)):
            word_in_file = lines_in_word_file[i].split()
            tag_in_file = lines_in_tag_file[i].split()
            length = min(len(word_in_file), len(tag_in_file))
            #Create the word_tag pair
            sent_len = 0
            for j in range(length):
                if anago:
                    valid_file.write(str(word_in_file[j])+'\t'+tag_in_file[j]+'\n')
                if guilla:
                    valid_file.write(str(word_in_file[j])+' '+tag_in_file[j]+'\n')
                sent_len = sent_len + 1
                #if sent_len == max_sentence_length:
                    #sent_len = 0
                if word_in_file[j] == '.' and tag_in_file[j] == 'O':
                    valid_file.write('\n')
valid_file.close()


test_file = open(test_file_path, 'w')

for f in test_file_list:
    word_file_path = os.path.join(doc_dir, f)
    tag_file_path = os.path.join(tag_dir, f)
    lines_in_word_file = []
    lines_in_tag_file = []
    with open(word_file_path, "rt") as word_file:
        for line in word_file:
            lines_in_word_file.append(line)
    with open(tag_file_path, "rt") as tag_file:
        for line in tag_file:
            lines_in_tag_file.append(line)
    if (len(lines_in_word_file) == len(lines_in_tag_file)) and len(lines_in_word_file) > 0:
        for i in range(len(lines_in_word_file)):
            word_in_file = lines_in_word_file[i].split()
            tag_in_file = lines_in_tag_file[i].split()
            length = min(len(word_in_file), len(tag_in_file))
            #Create the word_tag pair
            sent_len = 0
            for j in range(length):
                if anago:
                    test_file.write(str(word_in_file[j])+'\t'+tag_in_file[j]+'\n')
                if guilla:
                    test_file.write(str(word_in_file[j])+' '+tag_in_file[j]+'\n')
                    
                sent_len = sent_len + 1
                #if sent_len == max_sentence_length:
                    #sent_len = 0
                if word_in_file[j] == '.' and tag_in_file[j] == 'O':
                    test_file.write('\n')
test_file.close()
```
